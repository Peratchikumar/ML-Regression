```python
#Importing the Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading the dataset
dataset=pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
#Categorical data converting into a nominal data
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
#Input\Output Split Method - #Independent - Input in the dataset
independent=dataset[['age','bmi','children','sex_male','smoker_yes']] 
independent
                     
```

```python
dependent=dataset[['charges']] #Dependent - Output in the dataset
dependent
```

```python
#Training set and Test Set Split Method
from sklearn.model_selection import train_test_split  #Sklearn Supports Machine Learning
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#StandardScaler for better results
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform (X_train)
X_test=sc.transform(X_test)
```

```python
#Model Creation
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid= {'kernel':['rbf','poly','sigmoid','linear'],
              'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}

grid=GridSearchCV(SVR(),param_grid, refit=True, verbose=3, n_jobs=-1)

#Fitting the model for grid search
grid.fit(X_train,y_train)
```

```python
#Print best parameter after tuning
#Print(grid.best_params_)
re=grid.cv_results_
print("The r_score value for best parameter{}:".format(grid.best_params_))
```

```python
#Table Creation
table=pd.DataFrame.from_dict(re)
table
```

```python
#Input

age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
#Future Prediction

future_prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_prediction={}".format(future_prediction))
```
